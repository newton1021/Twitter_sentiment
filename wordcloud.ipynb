```python
import numpy as np
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
#make function to get specific colors for word cloud
def cloudcolors(word=None, font_size=None, position=None, orientation=None, font_path=None, random_state=None):  
    #define the list of set colors  
    color_list = ["#FFFFFF", "#5C636B", "#509CE1"]  

    #return a random color in the list  
    return np.random.choice(color_list)
```

```python
testText = 'word, somthing, something, it, test test, test, tweets, twitter, I, words, fourteen, python, reading, read'

#make word cloud
wordcloud = WordCloud(background_color = "#15202b", color_func=cloudcolors).generate(testText)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

#save as png
wordcloud.to_file("web-page_template/twitter visuals/testcloud.png")
```
